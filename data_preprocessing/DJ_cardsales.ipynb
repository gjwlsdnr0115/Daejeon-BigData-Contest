```python
sumyh.reset_index(level=0, inplace=True)
```

```python
merge = pd.merge(df,sumyh, how='left')
merge.tail()
```

```python
yil_card=card[lambda x: (x['분류'] == '유흥') | (x['분류'] =='일반음식점')]
yil_card.head()
```

```python
grouped = yil_card['매출액'].groupby(yil_card['행정동'])
sumyil = grouped.sum().to_frame()
sumyil.head()
```

```python
sumyil.sort_values('매출액',ascending=False).head(10)
```

```python
sumyil.reset_index(level=0, inplace=True)
merge2 = pd.merge(df,sumyil, how='left')
merge2.tail()
```

```python
with pd.ExcelWriter('동별유흥매출액.xlsx') as writer:  
    merge.to_excel(writer, sheet_name='유흥합계')
    merge2.to_excel(writer, sheet_name='유흥+일반음식점합계')
```
