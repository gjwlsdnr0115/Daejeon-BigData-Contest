```python
import pandas as pd
import numpy as np
import requests
```

```python
# import cctv data
df = pd.read_csv('서구_CCTV.csv', encoding='utf-8')
pd.set_option('display.max_rows', df.shape[0]+1)
```

```python
cctv = df[['소재지도로명주소', '소재지지번주소', '카메라대수']]
```

```python
cctv.head()
```

```python
# 도로명주소가 없는 데이터는 지번주소로 대체
cctv['소재지도로명주소'] = cctv['소재지도로명주소'].fillna(cctv['소재지지번주소'])
```

```python
cctv.info()
```

```python
# 카카오 행정동 검색 api
url = "https://dapi.kakao.com/v2/local/search/address.json?"
apikey = "2bac3422318896e96bfb6f04e97bdaaa"
```

```python
# api 사용하여 각 데이터 별 행정동 주소 추출
dlist=[]
for add in cctv['소재지도로명주소']:
    req= requests.get(url, params={'query': add},
                     headers={'Authorization': 'KakaoAK ' + apikey})
    js= req.json()
    print(js)
    if js['documents'] == []:
        dlist.append('not found')
    else:
        if js['documents'][0]['address'] == None:
            dong = js['documents'][0]['road_address']['region_3depth_name']
            dlist.append(dong)
        else:
            dong = js['documents'][0]['address']['region_3depth_h_name']
            dlist.append(dong)
```

```python
cctv['행정동'] = dlist
```

행정동이 검색되지 않는 도로명주소는 직접 검색 후 데이터 입력하였음

```python
cctv['행정동'].unique()
```

```python
cams = cctv.groupby('행정동')['카메라대수'].sum()
```
