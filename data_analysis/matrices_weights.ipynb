```python
import pandas as pd
import numpy as np
```

```python
# 전체 데이터 불러오기
merged_df = pd.read_csv('re.csv', encoding='utf-8')
pd.set_option('display.max_rows', merged_df.shape[0]+1)
merged_df.drop('Unnamed: 0', axis = 1, inplace=True)
```

```python
merged_df.head()
```

```python
# 각 행정동으로 이루어진 행렬 생성
dongs = []
for dong in merged_df['행정동']:
    dongs.append(dong)

m = pd.DataFrame(columns=dongs, index=dongs)

# 행정동을 index로 지정
merged_df.set_index('행정동', inplace=True)
```

```python
m.head()
```

```python
# 각 행정동의 상대적 값을 구하는 함수
def divide(data, c):
    n = data / c
    return n
```

```python
# 각 행정동 divide함수 적용하여 행렬 값 입력
for dong in dongs:
    m[dong] = merged_df.apply(lambda x: divide(x['cctv_re'], merged_df.at[dong, 'cctv_re']), axis=1)
```

```python
m.head()
```

```python
# 행렬 제곱 후 상대적 중요도 산출
m_dot = m.dot(m)
m_dot['sum'] = m_dot.sum(axis=1)
total = m_dot['sum'].sum()
m_dot['weight'] = m_dot['sum'].apply(lambda x: x / total)  # 상대적 값 계산
m_dot.head()
```

```python
# 모든 중요도의 합 = 1
check = m_dot['weight'].sum()
check
```

```python
final = m_dot[['sum', 'weight']]
final
```

```python
final.to_csv('weight_cctc_re.csv')
```
